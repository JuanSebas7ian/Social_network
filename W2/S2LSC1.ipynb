Carguemos el archivo de red mediante *Pandas*, guardando la informacin en ``nodos`` y en ``arcos``.

```python
# Nodos y arcos de la red
nodos = pd.read_csv('NodosCL.csv', sep=',', index_col="Id")
arcos = pd.read_csv('ArcosCL.csv', sep=',', index_col="Id")
```

Los archivos poseen informacin que no es necesaria asi que podemos remover esas columnas.

```python
# Limpiemos las bases
nodos = nodos.drop(columns=['timeset','username','post_activity','fan_count','talking_about_count','users_can_post','link'])
arcos = arcos.drop(columns=['Label','timeset'])
```

Con los datos ya cargados, pasemos a construir la red ``G``, como una red **dirigida**.

```python
# Construir red
G=nx.DiGraph()

# Nodos
# Agregar los nodos por su identificador
G.add_nodes_from(nodos.index)
# Agregar los atributos de los nodos
# Primero recorrer los nodos que se agregaron a la red
for nodo in G.nodes:
    # En cada nodo recorremos los atributos de los nodos
    for atr in nodos:
        # Agregar cada atributo al nodo
        G.nodes[nodo][atr]=nodos[atr][nodo]
        
# Arcos
# Agregar los arcos
G.add_edges_from([tuple(x) for x in arcos[['Source','Target']].to_numpy()])

## Nota: para este ejercicio no requerimos los pesos dado que los pesos son siempre uno (1)
```

### Grado de un nodo

El *grado* de un nodo es el nmero de arcos que salen y/o llegan de este.<br>
Existen tres formas de medir el grado de un *nodo*:
1. ``.in_degree(nodo)``, que mide el nmero de arcos que llegan a un nodo determinado (*nota:* solo vlido para redes dirigidas).
2. ``.out_degree(nodo)``, que mide el nmero de arcos que salen de un nodo determinado (*nota:* solo vlido para redes dirigidas).
3. ``.degree(nodo)``, que mide el nmero total de arcos de un nodo (*nota:* vlido para redes dirigidas y no dirigidas).

### Centralidad Eigenvector

La centralidad eigenvector calcula la centralidad de un nodo en funcin de la centralidad de sus vecinos. La centralidad eigenvector para el nodo $i$ es el $i$-simo elemento del vector definido por la ecuacin:
$$Ax=\lambda x$$
donde $A$ es la matriz de adyacencia de la red con valor propio $\lambda$. En virtud del teorema de Perron-Frobenius, existe una solucin nica $x$, donde todas sus entradas son positivas, si $\lambda$ es el valor propio ms grande de la matriz de adyacencia $A$.<br>

*Mark E. J. Newman. Networks: An Introduction. Oxford University Press, USA, 2010, pp. 169.*

Para encontrar la centralidad eigenvector de cada nodo se utiliza la funcin ``nx.eigenvector_centrality(G)``, la cul regresa un diccionario donde las llaves son el *id* de los nodos y los valores son la centralidad eigenvector.

```python
nx.eigenvector_centrality(G)
```

<a id = "betweenness"></a>
### Betweenness

La centralidad de intermediacin (betweenness) de un nodo $v$ es la suma de la fraccin de los todos los caminos ms cortos entre todos los pares que pasan por $v$.
$$c_B(v) =\sum_{s,t \in V} \frac{\sigma(s, t|v)}{\sigma(s, t)}$$
Donde $V$ es el set de nodos, $\sigma(s, t)$ es el nmero de caminos ms cortos entre $(s, t)$, y $\sigma(s, t|v)$ es el nmero de esos caminos que pasan por un nodo $v$ que no sean $s,t$. Si $s = t$, $\sigma(s, t) = 1$,y si $v \in {s, t}$,$\sigma(s, t|v) = 0$. <br>

*Ulrik Brandes: On Variants of Shortest-Path Betweenness Centrality and their Generic Computation. Social Networks 30(2):136-145, 2008. https://doi.org/10.1016/j.socnet.2007.11.001*

Para encontrar la centralidad de intermediacin de cada nodo se utiliza la funcin ``nx.betweenness_centrality(G)``, la cul regresa un diccionario donde las llaves son el *id* de los nodos y los valores son la centralidad de intermediacin.

```python
nx.betweenness_centrality(G)
```

<a id = "closeness"></a>
### Closeness

La centralidad de proximidad (closeness) de un nodo $u$ es el recproco de la distancia de camino ms corto promedio a $u$ en todos los $n-1$ nodos alcanzables.
$$C(u) = \frac{n - 1}{\sum_{v=1}^{n-1} d(v, u)}$$
donde $d(v, u)$ es la distancia de camino ms corta entre $v$ y $u$, y $n$ es el nmero de nodos que pueden llegar a $u$. 

Observe que los valores ms altos de cercana indican una mayor centralidad.

*Linton C. Freeman: Centrality in networks: I. Conceptual clarification. Social Networks 1:215-239, 1979. https://doi.org/10.1016/0378-8733(78)90021-7*

Para encontrar la centralidad de proximidad de cada nodo se utiliza la funcin ``nx.closeness_centrality(G)``, la cul regresa un diccionario donde las llaves son el *id* de los nodos y los valores son la centralidad de proximidad.

```python
nx.closeness_centrality(G)
```

### Clustering

Para grafos no dirigidos, el clustering de un nodo $u$ es la fraccin de tringulos posibles a travs de ese nodo que existen,

$$c_u = \frac{2 T(u)}{deg(u)(deg(u)-1)},$$

donde $T(u)$ es el nmero de tringulos a travs del nodo $u$ y $deg(u)$ es el grado de $u$.

Para encontrar el clustering de cada nodo se utiliza la funcin ``nx.clustering(G)``, la cul regresa un diccionario donde las llaves son el *id* de los nodos y los valores son el clustering.

```python
nx.clustering(G)
```

<a id = "diametro"></a>
### Diametro de la red

El diametro de la red esta definido como la distancia ms larga entre las distancias ms cortas.

Antes de calcular el diametro es importante comprender tres conceptos importantes de las redes:
1. Una red est **conectada** si es posible ir de cualquier nodo $i$ a cualquier nodo $j$.
2. Una red est **fuertemente conectada** si es posible ir de cualquier nodo $i$ a cualquier nodo $j$ respetando la direccin de los arcos.
3. Una red est **debilmente conectada** si es posible ir de cualquier nodo $i$ a cualquier nodo $j$ sin respetar la direccin de los arcos, pero no se puede ir de cualquier nodo $i$ a cualquier nodo $j$ respetando la direccin de los arcos.

Para calcular el diametro es necesario que la red este fuertemente conectada, es decir se pueda alcanzar todos los nodos desde cualquier nodo. Podemos validar que la red est fuertemente conectada mediante la funcin ``nx.is_strongly_connected(G)``, la cual regresa un booleano validando si la red est fuertemente conectada o no.

```python
# Verificar que el grafo est fuertemente conectado
nx.is_strongly_connected(G)
```

Ahora, como la red en la que se que se est trabajando es dirigida, significa que al seguir el sentido de los arcos no se puede llegar a todos los nodos, por lo que la red no esta fuertemente conectada. Para corregir esto y calcular el diametro podemos transformar la red en un no dirigida usando la funcin ``G.to_undirected()``, y guardando la red resultante en ``H``.

```python
# Convirtamos el grafo en uno no dirigido
H = G.to_undirected()
```

Una vez convertida en no dirigiada, validemos que la red est completamente conectada, para lo cual usamos la funcin ``nx.is_connected(H)``, que regresa un booleano confirmando si la red esta o no conectada.

```python
# El grado est conectado
nx.is_connected(H)
```

Dado que la red est conectada y es no dirigida podemos calcular todas las distancias ms cortas entre todo par de nodos en la red. Para esto podemos usar la funcin ``nx.all_pairs_shortest_path(H)``, y guardar el generador de todas las rutas entre el origen y el destino en un diccionario.

```python
# Calculemos todo las distancias ms cortas entre todos los nodos
path = dict(nx.all_pairs_shortest_path(H))
```

Con los caminos ms cortos encontrados busquemos el ms largo.

```python
#Encontremos la distancia ms larga
mayor_distancia=0
for i in H.nodes():
    for j in H.nodes():
        if i != j:
            if len(path[i][j]) > mayor_distancia:
                mayor_distancia=len(path[i][j])-1
                
print('El diametro de la red es', mayor_distancia)
```

Ahora sabemso que el diametro de la red es *4*.

Networkx, nos ayuda en este proceso y podemos encontrar facilmente el diametro de la red con la funcin ``nx.diameter(H)``.

```python
# Por medio de las funciones de Networkx
nx.diameter(H)
```

*Resultado esperado:* ``4``.

<a id = "densidad"></a>
### Densidad de la red

La densidad para grafos no dirigidos est definida como:

$$d = \frac{2m}{n(n-1)}$$

y para grafos dirigidos es:

$$d = \frac{m}{n(n-1)}$$

donde $n$ es el nmero de nodos y $m$ es el nmero de aristas en la red. En otras palabras, la densidad nos dice que propocin de conexiones posibles entre los nodos de la red  existen.

Para calcular la densidad de la red podemos usar la frmula.

```python
# No olviden que es una red dirigida
densidad=nx.number_of_edges(G)/(nx.number_of_nodes(G)*(nx.number_of_nodes(G)-1))
densidad
```

*Resultado esperado:* ``0.01167``.

Tambin podemos usar la funcin de NetworkX ``nx.density(G)``.

```python
# Funcin de NetworkX
nx.density(G)
```

*Resultado esperado:* ``0.01167``.

Esto nos dice que de todas las conexiones posibles en la red solo el 1.1% de estas existe en la red.

<a id = "reciprocidad"></a>
### Reciprocidad de la red

La reciprocidad de una red dirigida se define como el porcentaje de relaciones que van en ambos sentidos entre todo par de nodos conectados en la red. Formalmente,

$$r = \frac{|{(u,v) \in G|(v,u) \in G}|}{|{(u,v) \in G}|}$$

Usemos la funcin ``nx.reciprocity(G)``.

```python
nx.reciprocity(G)
```

*Resultado esperado:* ``0.3215``.

Esto nos dice que de todas las conexiones que hay en la red el 32.15% de estas son reciprocas o bidireccionales.

<a id = "asortatividad"></a>
### Asortatividad

El coeficiente de asortividad es el coeficiente de correlacin de Pearson de grado entre pares de nodos conectados. Los valores positivos de r indican una correlacin entre nodos de grado similar, mientras que los valores negativos indican relaciones entre nodos de diferente grado.<br>

En otras palabras, la asociatividad mide la similitud de las conexiones en el grafo con respecto al grado del nodo.<br>

*M. E. J. Newman, Mixing patterns in networks, Physical Review E, 67 026126, 2003*

Usamos la funcin ``nx.degree_assortativity_coefficient(G)``.

```python
nx.degree_assortativity_coefficient(G)
```

*Resultado esperado:* ``-0.1309``.
